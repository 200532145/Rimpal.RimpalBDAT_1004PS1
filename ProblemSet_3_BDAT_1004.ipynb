#### Step 4. Discover what is the mean age per occupation
#### Step 5. Discover the Male ratio per occupation and sort it from the most to the least
#### Step 6. For each occupation, calculate the minimum and maximum ages
#### Step 7. For each combination of occupation and sex, calculate the mean age
#### Step 8. For each occupation present the percentage of women and men

## Answer 1: -
************

```python
# Step 1 : Import the neccessary file
import numpy as np
import csv
import requests
import pandas as pd
```

```python
# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|') # Step 3: Assign the dataset to the variable 'users'
```

```python

users.head()
```

```python
# Step 4: Calculate the mean age per occupation
mean_age_by_occupation = users.groupby('occupation')['age'].mean()
print("Ans 1.")
print("Mean Age per Occupation:")
print(mean_age_by_occupation)
```

```python
# Step 5: Calculate the Male ratio per occupation and sort it
users['is_male'] = (users['gender'] == 'M').astype(int)
male_ratio_per_occupation = users.groupby('occupation')['is_male'].mean().sort_values(ascending=False)
print("\n Ans 2.")
print("\nMale Ratio per Occupation (Most to Least):")
print(male_ratio_per_occupation)
```

```python
# Step 6: Calculate the minimum and maximum ages per occupation
min_age_by_occupation = users.groupby('occupation')['age'].min()
max_age_by_occupation = users.groupby('occupation')['age'].max()
print("\n Ans 3.")
print("\nMinimum Age per Occupation:")
print(min_age_by_occupation)

print("\n Ans 4.")
print("Maximum Age per Occupation:")
print(max_age_by_occupation)
```

```python
# Step 7: Calculate the mean age per combination of occupation and sex
mean_age_by_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("\n Ans 5.")
print("\nMean Age per Occupation and Sex:")
print(mean_age_by_occupation_sex)
```

```python
# Step 8: Calculate the percentage of women and men per occupation
total_by_occupation = users.groupby('occupation')['gender'].count()
percentage_women_per_occupation = (users[users['gender'] == 'F'].groupby('occupation')['gender'].count() / total_by_occupation) * 100
percentage_men_per_occupation = 100 - percentage_women_per_occupation
print("\n Ans 6.")
print("\nPercentage of Women per Occupation:")
print(percentage_women_per_occupation)

print("\n Ans 7.")
print("\nPercentage of Men per Occupation:")
print(percentage_men_per_occupation)
```

```python
import pandas as pd
import numpy as np

# Step 2: Import the dataset
data = pd.read_csv('wind.txt', sep='\s+', parse_dates=[[0, 1, 2]])

# Step 4: Fix year 2061
def fix_year(year):
    if year > 2000:
        year -= 100
    return year

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(lambda x: x.replace(year=fix_year(x.year)))

# Step 5: Set datetime index
data.set_index('Yr_Mo_Dy', inplace=True)

# Step 6: Count missing values by location
missing_values = data.isnull().sum()

# Step 7: Count non-missing values
non_missing_values = data.count().sum()

# Step 8: Calculate mean windspeed
mean_windspeed = data.mean().mean()

# Step 9: Calculate location statistics
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

# Step 10: Calculate daily statistics
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])

# Step 11: Calculate average windspeed in January for each location
#january_avg = data[data.index.month == 1].groupby([data.index.year, data.index.month]).mean()
jan_avg = data[data.index.month == 1].mean()


# Step 12: Downsample to yearly frequency
yearly_data = data.resample('Y').mean()

# Step 13: Downsample to monthly frequency
monthly_data = data.resample('M').mean()

# Step 14: Downsample to weekly frequency
weekly_data = data.resample('W').mean()

# Step 15: Calculate weekly statistics for the first 52 weeks
weekly_stats = weekly_data['1961-01-02':'1961-12-25'].agg(['min', 'max', 'mean', 'std'])
print("Ans 1.")
print("Missing values count for each location:\n", missing_values)
print("\nAns 2.")
print("\nNon-missing values count in total:", non_missing_values)
print("\nAns 3.")
print("\nMean windspeed over all locations and times:", mean_windspeed)
print("\nAns 4.")
print("\nloc_stats DataFrame:\n", loc_stats)
print("\nAns 5.")
print("\nday_stats DataFrame:\n", day_stats)
print("\nAns 6.")
print("\nAverage windspeed in January for each location:\n", jan_avg)
print("\nAns 7.")
print("\nYearly data:\n", yearly_data)
print("\nAns 8.")
print("\nMonthly data:\n", monthly_data)
print("\nAns 9.")
print("\nWeekly data:\n", weekly_data)
print("\nAns 10.")
print("\nWeekly statistics:\n", weekly_stats.head(52))
```

```python
# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t') # Step 3: Assign the dataset to the variable 'chipo'
```

```python
# Step 4: See the first 10 entries
first_10_entries = chipo.head(10)
first_10_entries.head(10)
```

```python
# Step 5: Number of observations in the dataset
num_observations = chipo.shape[0]
print("\nNumber of observations in the dataset:", num_observations)
```

```python
# Step 6: Number of columns in the dataset
num_columns = chipo.shape[1]
print("Number of columns in the dataset:", num_columns)
```

```python
# Step 7: Print the name of all the columns
columns_names = chipo.columns

# Step 8: How the dataset is indexed
index_type = chipo.index

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()

# Step 10: Number of items ordered for the most-ordered item
most_ordered_item_count = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()

# Step 11: Most ordered item in the 'choice_description' column
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()

# Step 12: Total number of items ordered
total_items_ordered = chipo['quantity'].sum()

# Step 13: Convert item price to a float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Step 14: Calculate revenue
revenue = (chipo['item_price'] * chipo['quantity']).sum()

# Step 15: Number of orders
num_orders = chipo['order_id'].nunique()

# Step 16: Average revenue per order
avg_revenue_per_order = revenue / num_orders

# Step 17: Number of different items sold
num_different_items = chipo['item_name'].nunique()

print("\nstep 7.")
print("Columns names:", columns_names)
print("\nstep 8.")
print("Index type:", index_type)
print("\nstep 9.")
print("Most ordered item:", most_ordered_item)
print("\nstep 10.")
print("Number of items ordered for the most-ordered item:", most_ordered_item_count)
print("\nstep 11.")
print("Most ordered choice description:", most_ordered_choice)
print("\nstep 12.")
print("Total number of items ordered:", total_items_ordered)
print("\nstep 13.")
print("Item price type after conversion:", chipo['item_price'].dtype)
print("\nstep 14.")
print("Total revenue:", revenue)
print("\nstep 15.")
print("Number of orders:", num_orders)
print("\nstep 16.")
print("Average revenue per order:", avg_revenue_per_order)
print("\nstep 17.")
print("Number of different items sold:", num_different_items)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from local CSV file
data_file_path = "C:/Users/Rimpal anttal/Downloads/us-marriages-divorces-1867-2014.csv" 
read_data = pd.read_csv(data_file_path)

# Filter data for the years between 1867 and 2014
filtered_data = read_data[(read_data['Year'] >= 1867) & (read_data['Year'] <= 2014)]

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(filtered_data['Year'], filtered_data['Marriages_per_1000'], label='Marriages per capita')
plt.plot(filtered_data['Year'], filtered_data['Divorces_per_1000'], label='Divorces per capita')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()
plt.tight_layout()


plt.show()
```
